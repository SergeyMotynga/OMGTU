```python
# Задание 1
import pandas as pd

data = {
    'Name': [
        'Strom, Mrs. Wilhelm (Elna Matilda Persson)',
        'Navratil, Mr. Michel ("Louis M Hoffman")',
        'Minahan, Miss. Daisy E'
    ],
    'Age': [29, 36.5, 33],
    'Sex': ['female', 'male', 'female']
}
df1 = pd.DataFrame(data)
df1
```

```python
# Задание 2
df2 = pd.read_csv("C:/Users/serega/Downloads/titanic_csv.csv", sep=';')
df2
```

```python
# Задание 3
df3 = pd.read_csv("https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt", sep=',')
df3
```

```python
# Задание 4
df2.columns = df2.columns.str.upper()
df3.columns = df3.columns.str.upper()

concatDf = pd.concat([df2, df3])

concatDf = concatDf.drop_duplicates()

concatDf
```

```python
# Задание 5
concatDf = concatDf.set_index('PASSENGERID').sort_values(by='PASSENGERID')
concatDf
```

```python
# Задание 6
print("Информация о датафрейме:")
print(concatDf.info())
```

```python
# Задание 6
print("Статистика по датафрейму:")
print(concatDf.describe())
```

```python
# Задание 7
temp = concatDf.iloc[0].copy()
concatDf.loc[1] = concatDf.loc[2]
concatDf.iloc[1] = temp
concatDf
```

```python
temp = concatDf.loc[1].copy()
concatDf.loc[1] = concatDf.loc[2]
concatDf.loc[2] = temp
concatDf
```

```python
# Задание 8
concatDf['SEX'] = concatDf['SEX'].replace({'male': 'M', 'female': 'F'})
concatDf['SEX']
```

```python
# Задание 9
ticketCount = concatDf.groupby('TICKET').count()
ticketCountSixOrMoreThanSix = ticketCount[ticketCount['NAME'] >= 6]
passengers_list = concatDf[concatDf['TICKET'].isin(ticketCountSixOrMoreThanSix.index)]
passengers_list
```

```python
# Задание 10
cabins_df1 = concatDf[concatDf['NAME'].isin(df1['Name'])]
cabin_mates = concatDf[concatDf['CABIN'].isin(cabins_df1['CABIN'])]
cabin_mates
```

```python
# Задание 11
concatDf['BIRTHYEAR'] = (1912 - concatDf['AGE']).fillna(0).astype(int)
concatDf['BIRTHYEAR']
```

```python
# Задание 12
concatDf['COMPANION'] = (concatDf.groupby('CABIN')['NAME'].transform('count') - 1).fillna(0).astype('int64')

concatDf[['NAME', 'COMPANION']]
```

```python
# Задание 13
# concatDf.iloc[[0, 1]] = concatDf.iloc[[1, 0]].values

# concatDf.head()
temp = concatDf.iloc[0].copy()
concatDf.iloc[0] = concatDf.iloc[1]
concatDf.iloc[1] = temp
concatDf
```

```python
# Задание 14
concatDf.to_csv('titanic_csv.csv', sep=';')
```

```python
# Задание 15
top_10 = concatDf.sort_values('FARE', ascending=False).head(10)
print(top_10[['NAME', 'FARE', 'PCLASS']].to_string())
```

```python
# Задание 16
survival_by_sex = concatDf.groupby(['SEX', 'SURVIVED'])['SEX'].count().unstack()
survival_by_sex.columns = ['Не выжили', 'Выжили']
print(survival_by_sex)
```

```python
# Задание 17
survival_by_class = concatDf.groupby(['PCLASS', 'SURVIVED'])['PCLASS'].count().unstack()
survival_by_class.columns = ['Не выжили', 'Выжили']
print(survival_by_class)
```
